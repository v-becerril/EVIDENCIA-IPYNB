```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
#convertir data frame
df = pd.read_csv("airbnb_madrid_filtered.csv")
df
```

```python
# Seleccionamos 5 nombres de anfitriones distintos (no nulos)
hosts_muestra = df["nombre_anfitrion"].dropna().unique()[:5]
print("Hosts seleccionados:", hosts_muestra)

# Filtramos el dataframe para quedarnos con esos 5
df_hosts = df[df["nombre_anfitrion"].isin(hosts_muestra)]

# Mostramos una vista rápida
df_hosts.head(10)
```

```python
#b) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
hosts_2020 = df[df["anfitrion_desde"] > "2020-12-31"]
hosts_2020
```

```python
#c) Los registros de los host que responden “a más tardar en 1 día”
prim_col = "tiempo_respuesta_anfitrion"

mask = df[prim_col].str.contains("within a day", na=False)

ordenarcols = [prim_col] + [col for col in df.columns if col != prim_col]
RespondenMax1 = df[mask][ordenarcols]
RespondenMax1
```

```python
# d) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
pos = range(399, len(df), 200)             
out = df.iloc[pos].copy()

# Insertar una columna con el número de fila 
row_nums = [p + 1 for p in pos]            
out.insert(0, "row_num", row_nums)

print(out.shape)
out.head()
```

```python
#e) Los registros de los tipos de cuarto: “Entire home” y “Private room”
col = "tipo_habitacion"
vals = ["Entire home", "Private room"]
df_rooms = df.loc[df[col].isin(vals), [col] + [c for c in df.columns if c != col]]
df_rooms
```

```python
#f) Los registros que “a lo mucho cuenten con 4 camas”
col = "camas"  
camas_4 = df.loc[df[col] <= 4, [col] + [c for c in df.columns if c != col]].dropna(subset=[col]).copy()
camas_4
```

```python
col = "es_superanfitrion"

superhost = df.loc[df[col] == "t", [col] + [c for c in df.columns if c != col]].dropna(subset=[col]).copy()
superhost
```

```python
#h) Los registros de las columnas “impares”
impares = df.iloc[:, ::2].copy()

print(impares.shape)
impares.head()
```
