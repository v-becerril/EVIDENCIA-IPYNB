```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
df= pd.read_csv('df_select_ROME_LIMPIO.csv') 
df
```

```python
df_Roma = df
df_Roma
```

**PROCEDIMIENTO PARA ELIMINAR VALORES ATÍPICOS EN DATAFRAME**


```python
#Corroboramos valores nulos del dataframe
display(df_Roma.isnull().sum())
```

```python
# --- DataFrame cualitativo ---
# Seleccionar columnas cualitativas
df_cualitativas = df.select_dtypes(include=['object']).copy()

# Agregar columnas id, host_id, latitude y longitude
df_cualitativas[['id', 'host_id', 'latitude', 'longitude', 'host_response_rate', 'host_acceptance_rate', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']] = df_Roma[['id', 'host_id', 'latitude', 'longitude', 'host_response_rate', 'host_acceptance_rate', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']]

# Reordenar para que id y host_id sean las primeras columnas
cols = ['id', 'host_id'] + [col for col in df_cualitativas.columns if col not in ['id', 'host_id']]
df_cualitativas = df_cualitativas[cols]

# --- DataFrame cuantitativo ---
df_cuantitativas = df.select_dtypes(include=['int64', 'float64']).drop(
    columns=['id', 'host_id', 'latitude', 'longitude', 'host_response_rate', 'host_acceptance_rate', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value'], errors='ignore')
df_cuantitativas = df_cuantitativas.drop(columns=['Unnamed: 0'])

df_cuantitativas
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
df_cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "IQR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df_cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
df_cuartiles= df_cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df_cuartiles
```

```python
#Corroboramos valores nulos (Atípicos) del dataframe LIMPIO
display(df_cuartiles.isnull().sum())
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean" para columnas que no importe decimales, y mediana a aquellas necesarias

#Realizamos una copia del dataframe
df_cuartiles_FINAL=df_cuartiles.copy()

# Lista de columnas que deben rellenarse con la MEDIANA
cols_mediana = ['host_listings_count', 'accommodates', 'bedrooms', 'beds',
    'minimum_nights', 'maximum_nights', 'price', 'number_of_reviews', 'availability_30', 'availability_60', 'availability_90', 'availability_365', 'calculated_host_listings_count']

# Rellenar con la MEDIANA las columnas específicas
df_cuartiles_FINAL[cols_mediana]=df_cuartiles_FINAL[cols_mediana].fillna(round(df_cuartiles[cols_mediana].median(),1))

```

```python
# Lista de columnas que deben rellenarse con la MEDIA
cols_media = [ 'reviews_per_month']

# Rellenar con la MEDIA las columnas específicas
df_cuartiles_FINAL[cols_media]=df_cuartiles_FINAL[cols_media].fillna(round(df_cuartiles[cols_media].median(),1))

```

```python
#Corroboramos valores nulos (Atípicos) del dataframe FINAL
display(df_cuartiles_FINAL.isnull().sum())
```

```python
# Unimos el dataframe cuantitativo limpio con el dataframe cualitativo

Dataset_libre_Outliers = pd.concat([df_cualitativas, df_cuartiles_FINAL], axis=1)
pd.set_option('display.max_columns', None)

Dataset_libre_Outliers
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Dataset_libre_Outliers.isnull().sum()
valores_nulos
```

```python
DF_ROMA_NO_FINAL = Dataset_libre_Outliers
DF_ROMA_NO_FINAL
```

```python
#Convertir DataFrame a CSV (Cuartiles)
DF_ROMA_NO_FINAL.to_csv("ROMA FINAL.csv", index = False)
```
