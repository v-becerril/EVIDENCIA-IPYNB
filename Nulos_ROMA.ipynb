```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de archivo
df = pd.read_csv("df_select_ROME.csv")
df.head(5)
```

```python
#Identificar valores nulos por columna
display(df.isnull().sum())
```

```python
#Realizamos una copia del dataframe
df_ROMA = df
df_ROMA.head(5)
```

```python
#Establecer un nombre específico "Unknown" a todas los valores nulos dentro de las columnas 'host_name' y 'host_since'
df_ROMA["host_name"] = df_ROMA["host_name"].fillna("Unknown")
df_ROMA["host_since"] = df_ROMA["host_since"].fillna("00/00/00")

display(df_ROMA.isnull().sum())
```

```python
## LLenamos los NAs de la columna de 'host_response_time' con el string más repetido; "within an hour"
valor_mas_repetido_response = df_ROMA["host_response_time"].mode()[0]
print("El valor más repetido es:", valor_mas_repetido_response)

df_ROMA["host_response_time"] = df_ROMA["host_response_time"].fillna(valor_mas_repetido_response)

display(df_ROMA.isnull().sum())
df_ROMA
```

```python
#Eliminar un signo de columna 'host_response_rate' en dataframe df (copia)
df_ROMA['host_response_rate']=df_ROMA['host_response_rate'].str.replace('%', '')
df_ROMA
```

```python
#Eliminar un signo de columna 'host_response_rate' en dataframe df (original)
df['host_response_rate']=df['host_response_rate'].str.replace('%', '')
df
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df_ROMA['host_response_rate']= df_ROMA['host_response_rate'].astype(float)
df['host_response_rate']= df['host_response_rate'].astype(float)
```

```python
#Sustituir valores nulos con promedio o media en columna 'host_response_rate'
df_ROMA["host_response_rate"]=df_ROMA["host_response_rate"].fillna(round(df["host_response_rate"].mean(),1))
display(df_ROMA.isnull().sum())
```

```python
#Eliminar un signo de columna 'host_acceptance_rate' en dataframe df (copia)
df_ROMA['host_acceptance_rate']=df_ROMA['host_acceptance_rate'].str.replace('%', '')
df_ROMA

#Eliminar un signo de columna 'host_acceptance_rate' en dataframe df (original)
df['host_acceptance_rate']=df['host_acceptance_rate'].str.replace('%', '')
df
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df_ROMA['host_acceptance_rate']= df_ROMA['host_acceptance_rate'].astype(float)
df['host_acceptance_rate']= df['host_acceptance_rate'].astype(float)
print(df.dtypes)
```

```python
#Sustituir valores nulos con promedio o media en columna 'host_acceptance_rate'
df_ROMA["host_acceptance_rate"]=df_ROMA["host_acceptance_rate"].fillna(round(df["host_acceptance_rate"].mean(),1))
display(df_ROMA.isnull().sum())
```

```python
#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'host_is_superhost' y 'host_identity_verified'
df_ROMA["host_is_superhost"] = df_ROMA["host_is_superhost"].fillna("NONE")
df_ROMA["host_identity_verified"] = df_ROMA["host_identity_verified"].fillna("NONE")
display(df_ROMA.isnull().sum())
```

```python
#Establecer con media NAs de la columna 'host_listings_count'
df_ROMA["host_listings_count"]=df_ROMA["host_listings_count"].fillna(round(df["host_listings_count"].mean(),1))
display(df_ROMA.isnull().sum())
```

```python
#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'description'
df_ROMA["description"] = df_ROMA["description"].fillna("NONE")
display(df_ROMA.isnull().sum())
```

```python
## LLenamos los NAs de la columna de 'neighbourhood' con el vecindario más repetido; "Rome, Lazio, Italy"
valor_mas_repetido_response = df_ROMA["neighbourhood"].mode()[0]
print("El valor más repetido es:", valor_mas_repetido_response)

df_ROMA["neighbourhood"] = df_ROMA["neighbourhood"].fillna(valor_mas_repetido_response)

display(df_ROMA.isnull().sum())
```

```python
#Establecer con media NAs de la columna 'bedrooms' y 'beds'
df_ROMA["bedrooms"]=df_ROMA["bedrooms"].fillna(round(df["bedrooms"].mean(),1))
df_ROMA["beds"]=df_ROMA["beds"].fillna(round(df["beds"].mean(),1))
display(df_ROMA.isnull().sum())
```

```python
#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'bathrooms_text '
df_ROMA["bathrooms_text"] = df_ROMA["bathrooms_text"].fillna("NONE")
display(df_ROMA.isnull().sum())
```

```python
# Convertir la columna 'price' a numérica (eliminando símbolos de $ y comas)
df["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Rellenar los NaN en 'price' con el promedio del room_type correspondiente
df["price"] = df.apply(
    lambda row: promedios[row["room_type"]] if pd.isna(row["price"]) else row["price"], axis=1)
display(df_ROMA.isnull().sum())
```
