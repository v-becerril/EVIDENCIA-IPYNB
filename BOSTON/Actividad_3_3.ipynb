```python
import pandas as pd
import numpy as np
df_entendimiento = pd.read_csv("/Users/hector/Desktop/Actividad 3.1, 3.2 y 3.3 (Valores Nulos)/BOSTON 50.csv")
print(df_entendimiento.shape)
df_entendimiento.head(10)
```

```python
#Ver cuantos valores nulos tiene
valores_nulos=df_entendimiento.isnull().sum()
valores_nulos
```

```python
# Host_response_time (531 nulos), sustitucipon de los nulos con "unknown"
df_entendimiento["host_response_time"] = df_entendimiento["host_response_time"].astype(str)
df_entendimiento["host_response_time"] = df_entendimiento["host_response_time"].replace("nan", "unknown")
```

```python
#Host_response_rate (531 nulos), sustituimos los valores faltantes directamente con 96% porque es la media
df_entendimiento["host_response_rate"] = np.where(pd.isnull(df_entendimiento["host_response_rate"]),"96%",df_entendimiento["host_response_rate"])

# Host_acceptance_rate (486 nulos)sustituimos los valores faltantes directamente con 89% porque es la media
df_entendimiento["host_acceptance_rate"] = np.where(pd.isnull(df_entendimiento["host_acceptance_rate"]),"89%",df_entendimiento["host_acceptance_rate"])
```

```python
#Host_is_superhost (176 nulos) sustituimos los valores faltantes directamente con "f" porque un nulo implica que no es superhost
df_entendimiento["host_is_superhost"] = np.where(
    pd.isnull(df_entendimiento["host_is_superhost"]), "f", df_entendimiento["host_is_superhost"]
)
```

```python
# Beds(834 nulos) sustituimos los valores faltantes con la mediana porque es más robusta frente a valores extremos
df_entendimiento["beds"] = df_entendimiento["beds"].fillna(df_entendimiento["beds"].median())

# Bedrooms(269 nulos) sustituimos los valores faltantes con la mediana porque es más representativa para variables discretas
df_entendimiento["bedrooms"] = df_entendimiento["bedrooms"].fillna(df_entendimiento["bedrooms"].median())

# Price (865 nulos) ---
# Limpiamos el símbolo de dólar y comas, luego convertimos a numérico
df_entendimiento["price"] = (
    df_entendimiento["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)


df_entendimiento["price"] = pd.to_numeric(df_entendimiento["price"], errors="coerce")


df_entendimiento["price"] = df_entendimiento["price"].fillna(df_entendimiento["price"].median())
```

```python
# Calendar update sustituimos los valores faltantes con "never" porque un nulo implica que el calendario nunca fue actualizado
df_entendimiento["calendar_updated"] = np.where(
    pd.isnull(df_entendimiento["calendar_updated"]),
    "never",
    df_entendimiento["calendar_updated"]
)
```

```python
# review_scores_ratingsustituimos los valores faltantes con la media porque es un indicador numérico y así mantenemos consistencia
df_entendimiento["review_scores_rating"] = df_entendimiento["review_scores_rating"].fillna(
    df_entendimiento["review_scores_rating"].mean()
)

# neighborhood_overview sustituimos los valores faltantes con "not provided" porque un nulo implica que el host no escribió descripción
df_entendimiento["neighborhood_overview"] = np.where(
    pd.isnull(df_entendimiento["neighborhood_overview"]),
    "not provided",
    df_entendimiento["neighborhood_overview"]
)
```

```python
# Description sustituimos los valores faltantes con "not provided" porque un nulo implica que el host no escribió descripción
df_entendimiento["description"] = np.where(
    pd.isnull(df_entendimiento["description"]),
    "not provided",
    df_entendimiento["description"]
)
```

```python
#Ver cuantos valores nulos tiene
valores_nulos=df_entendimiento.isnull().sum()
valores_nulos
```

```python
# Guardar como CSV sin índice
df_entendimiento.to_csv("airbnb_limpio.csv", index=False)
```
