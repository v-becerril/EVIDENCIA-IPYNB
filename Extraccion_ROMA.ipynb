```python
##Cargar archivo
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("listings_ROME.csv", on_bad_lines='skip')
df
```

```python
## a) Registros de 5 “host name” diferentes ---
five_hosts = df['host_name'].dropna().unique()[:5]
filtro_a = df[df['host_name'].isin(five_hosts)]
pd.set_option("display.max_columns", None)
filtro_a.to_csv("filtro_a.csv")
```

```python
## b) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
filtro_b = df[df['host_since'].dt.year > 2020]
pd.options.display.max_columns = None
filtro_b.to_csv("filtro_b.csv")
```

```python
## c) Los registros de los host que responden “a más tardar en 1 día”
filtro_c = df[df['host_response_time'] == "within a day"]
pd.options.display.max_columns = None
filtro_c.to_csv("filtro_c.csv")
```

```python
## d) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
filtro_d = df.iloc[::200, :]
pd.options.display.max_rows = None
filtro_d.to_csv("filtro_d.csv")
```

```python
## e) Los registros de los tipos de cuarto: “Entire home” y “Private room”
filtro_e = df[df['room_type'].isin(["Entire home/apt", "Private room"])]
pd.options.display.max_columns = None
filtro_e.to_csv("filtro_e.csv")
```

```python
## f) Los registros que “a lo mucho cuenten con 4 camas”
filtro_f = df[df['beds'] <= 4]
pd.options.display.max_columns = None
filtro_f.to_csv("filtro_f.csv")
```

```python
## g) Los registros que “sean superhost”
filtro_g = df[df['host_is_superhost'] == "t"]
pd.options.display.max_columns = None
filtro_g.to_csv("filtro_g.csv")
```

```python
## h) Los registros de las columnas “impares”
filtro_h = df.iloc[:, ::2]
filtro_h.to_csv("filtro_h.csv")
```
