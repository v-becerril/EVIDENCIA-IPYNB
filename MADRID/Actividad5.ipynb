```python
#Obtenemos el limite superior y el límite inferior de la columna objetivo
Max=df[['tasa_respuesta_anfitrion', 'tasa_aceptacion_anfitrion', 'total_anuncios_anfitrion', 'banos', 'camas', 'precio', 'promedio_max_noches', 'disponibilidad_365d', 'num_resenas', 'num_resenas_30d']].max()
Min=df[['tasa_respuesta_anfitrion', 'tasa_aceptacion_anfitrion', 'total_anuncios_anfitrion', 'banos', 'camas', 'precio', 'promedio_max_noches', 'disponibilidad_365d', 'num_resenas', 'num_resenas_30d']].min()
Limites= [Min, Max]
Limites
```

```python
#Calculamos el rango R
R=Max-Min
R
```

```python
#Calculamos el número de Intervalos de Clase "ni", aplicando la regla de Sturges
ni= 1+3.32*np.log10(n)
ni
```

```python
#Calculamos el Ancho del Intervalo "i"
i=R/ni
i
```
