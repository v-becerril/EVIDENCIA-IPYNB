```python
#HOST RESPONSE TIME

# Create a list of dataframes and their corresponding city names
dfs = {
    'Roma': dfRoma,
    'Boston': dfBoston,
    'Madrid': dfMadrid,
    'Bergamo': dfBergamo,
    'Mexico': dfMexico
}

# Rename the column in the Madrid DataFrame to match the others
Madrid = Madrid.rename(columns={'tiempo_respuesta_anfitrion': 'host_response_time'})

# Create an empty dictionary to store the comparative data
comparative_data = {}

# Iterate through the dataframes and extract the host_response_time data
for city, df in dfs.items():
    if 'host_response_time' in df.columns:
        # Replace 'Desconocida' and 'unknown' with a single category 'Unknown'
        df['host_response_time'] = df['host_response_time'].replace(['Desconocida', 'unknown'], 'Unknown')
        comparative_data[city] = df['host_response_time'].value_counts()

# Create a pandas DataFrame from the comparative data
comparative_table = pd.DataFrame(comparative_data)

# Fill NaN values with a hyphen for display
comparative_table = comparative_table.fillna('-')

# Display the comparative table
display(comparative_table)

# Iterate through the comparative table and create a bar chart for each city
for city in comparative_table.columns:
    plt.figure(figsize=(10, 6))
    # Get the data for the current city, excluding the hyphen values
    city_data = comparative_table[city][comparative_table[city] != '-'].sort_values(ascending=False) # Sort the data
    ax = sns.barplot(x=city_data.index, y=city_data.values, hue=city_data.index, palette='viridis', legend=False)
    plt.title(f'Host Response Time Distribution in {city}', fontsize=20) # Increased title size
    plt.xlabel('Host Response Time', fontsize=12) # Increased x-label size
    plt.ylabel('Total Count', fontsize=12) # Increased y-label size
    plt.xticks(rotation=45, ha='right', fontsize=15) # Increased x-tick label size
    plt.tight_layout()

    # Add percentages to the bars
    total = city_data.sum()
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2.
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10) # Increased annotation size

    plt.show()
```

```python
#HOST IS SUPERHOST

# Create an empty dictionary to store the comparative data for host_is_superhost
comparative_superhost_data = {}

# Define mapping for superhost status
superhost_mapping = {'t': 'Yes', 'Si': 'Yes', 'f': 'No', 'No': 'No', 'NONE': 'Unknown', 'No especificado': 'Unknown'}


# Iterate through the dataframes and extract the host_is_superhost data
for city, df in dfs.items():
    if city == 'Madrid' and 'es_superanfitrion' in df.columns:
        comparative_superhost_data[city] = df['es_superanfitrion'].map(superhost_mapping).value_counts()
    elif 'host_is_superhost' in df.columns:
        comparative_superhost_data[city] = df['host_is_superhost'].map(superhost_mapping).value_counts()


# Create a pandas DataFrame from the comparative superhost data
comparative_superhost_table = pd.DataFrame(comparative_superhost_data)

# Fill NaN values with a hyphen for display
comparative_superhost_table = comparative_superhost_table.fillna('-')

# Display the comparative superhost table
display(comparative_superhost_table)


# Iterate through the comparative superhost data and create a pie chart for each city
for city, data in comparative_superhost_data.items():
    plt.figure(figsize=(7, 7))
    plt.pie(data.values, labels=data.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 15})
    plt.title(f'Superhost Distribution in {city}', fontsize=25)
    plt.show()
```

```python
#NUMBER OF REVIEWS

# Create a list of dataframes and their corresponding city names
dfs = {
    'Roma': dfRoma,
    'Boston': dfBoston,
    'Madrid': dfMadrid,
    'Bergamo': dfBergamo,
    'Mexico': dfMexico
}

# Create an empty dictionary to store the comparative data for number_of_reviews
comparative_reviews_data = {}

# Define the ranges for number_of_reviews
review_ranges = [0, 10, 50, 100, float('inf')]
review_labels = ['0-9', '10-49', '50-99', '100+']

# Iterate through the dataframes and extract the number_of_reviews data
for city, df in dfs.items():
    if city == 'Madrid' and 'num_resenas' in df.columns:
        # Categorize 'num_resenas' into ranges
        df['review_range'] = pd.cut(df['num_resenas'], bins=review_ranges, labels=review_labels, right=False, include_lowest=True)
        comparative_reviews_data[city] = df['review_range'].value_counts().sort_index()
    elif city == 'Boston' and 'number_of_reviews_l30d' in df.columns:
         # Categorize 'number_of_reviews_l30d' into ranges for Boston
        df['review_range'] = pd.cut(df['number_of_reviews_l30d'], bins=review_ranges, labels=review_labels, right=False, include_lowest=True)
        comparative_reviews_data[city] = df['review_range'].value_counts().sort_index()
    elif 'number_of_reviews' in df.columns:
        # Categorize 'number_of_reviews' into ranges
        df['review_range'] = pd.cut(df['number_of_reviews'], bins=review_ranges, labels=review_labels, right=False, include_lowest=True)
        comparative_reviews_data[city] = df['review_range'].value_counts().sort_index()

# Define the desired order for the x-axis categories
order = ['0-9', '10-49', '50-99', '100+']

# Define a color palette for the cities
colors = sns.color_palette('viridis', len(comparative_reviews_data))

# Iterate through the comparative reviews data and create a bar plot for each city
for i, (city, data) in enumerate(comparative_reviews_data.items()):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=data.index, y=data.values, order=order, hue=data.index, palette='viridis', legend=False)
    plt.title(f'Number of Reviews Distribution in {city}')
    plt.xlabel('Number of Reviews')
    plt.ylabel('Total Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
